```python
import numpy as np
import pandas as pd
import io
d1 = pd.read_csv(io.BytesIO(uploaded['train.csv']))
d1.head(5)
```

```python
d2=d1.iloc[:,2:9]
d2.head()
```

```python
def one_hot_top_x(data,col_name,top_x):
  top=[x for x in d2[col_name].value_counts().head(10).index]
  for label in top:
    d2[col_name+"_"+label]=np.where(d2[col_name]==label,1,0)
```

```python
one_hot_top_x(d2,"X0",10)
one_hot_top_x(d2,"X1",10)
one_hot_top_x(d2,"X2",10)
one_hot_top_x(d2,"X3",10)
one_hot_top_x(d2,"X4",10)
one_hot_top_x(d2,"X5",10)
one_hot_top_x(d2,"X6",10)
```
